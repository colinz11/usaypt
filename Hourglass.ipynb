```python
import matplotlib.pyplot as plt
import math
import numpy as np
```

```python
volume = 0.00028
h_inital = 0.533
g = 9.8
h_final = 0.224
r = 0.02
theta = 0.488692
orfice = 0.0135
time_to_empty_funnel = 2
mass_of_lead = 1.795
F = mass_of_lead/time_to_empty_funnel

Q = volume/time_to_empty_funnel
rho = mass_of_lead/volume
v_inital = F/(rho*math.pi*orfice**2)
```

```python
time = []
mass3 = []
start_time = (math.sqrt(v_inital**2+2*g*h_inital)-v_inital)/g
end_time = (volume/Q)+(math.sqrt(v_inital**2+2*g*h_final)-v_inital)/g

time = np.arange(0, end_time, 0.001)
for t in time:
    if t < start_time:
        y=0
    else:
        y = rho*Q*(t-start_time)
    mass3.append(y)
```

```python
plt.plot(time,mass3)
plt.ylabel('Mass (kg)')
plt.xlabel('Time (s)')
plt.title('Mass (Region 3) vs Time')
```

```python
time = []
com3 = []

start_time = (math.sqrt(v_inital**2+2*g*h_inital)-v_inital)/g
end_time = (volume/Q)+(math.sqrt(v_inital**2+2*g*h_final)-v_inital)/g
print(start_time)
time = np.arange(0, end_time, 0.001)

for t in time:
    if t < start_time:
        y=0
    else:
        y = (Q*(t-start_time))/(2*math.pi * r**2)
    com3.append(y)
```

```python
plt.plot(time,com3,color='red')
plt.ylabel('Z Center of Mass (m)')
plt.xlabel('Time (s)')
plt.title('Center of Mass (Region 3) vs Time')
```

```python
time = []
mass1 = []
start_time = 0
time1 = time_to_empty_funnel
end_time = (volume/Q)+(math.sqrt(v_inital**2+2*g*h_final)-v_inital)/g
time = np.arange(start_time, end_time, 0.001)
for t in time:
    if t < time1:
        y = mass_of_lead - F*t
    else:
        y=0
    mass1.append(y)
```

```python
plt.plot(time,mass1)
plt.ylabel('Mass (kg)')
plt.xlabel('Time (s)')
plt.title('Mass (Region 1) vs Time')
```

```python
time = []
com1 = []
height_of_t = []
start_time = 0
time1 = time_to_empty_funnel
end_time = (volume/Q)+(math.sqrt(v_inital**2+2*g*h_final)-v_inital)/g
time = np.arange(start_time, end_time, 0.001)

for t in time:
    if t < time1:
        height = (((3*math.tan(theta))/(math.pi*rho)*(mass_of_lead-F*t)+(orfice)**3)**(1/3)-orfice)/math.tan(theta)
        height_of_t.append(height)
        mass_of_time =  mass_of_lead - F*t
        y = h_inital+((math.pi/2*rho*orfice**2*height**2)+
                      ((math.pi/3)*rho*orfice*math.tan(theta)*height**3)+
        ((math.pi/4)*rho*math.tan(theta)**2*height**4))/mass_of_time
    else:
        y = 0
    com1.append(y)

```

```python
time = []
mass2 = []
start_time = 0
end_time = (volume/Q)+(math.sqrt(v_inital**2+2*g*h_final)-v_inital)/g
print(end_time)
time_one = (math.sqrt(v_inital**2+2*g*h_inital)-v_inital)/g
time_two = volume/Q
print(time_one)
print(time_two)
h_bot = 0
h_top = 0
time = np.arange(start_time, end_time, 0.001)
for t in time:
    if t <= time_one:
        h_bot = h_inital-0.5*g*t**2-v_inital*t
        h_top = h_inital
    elif t <= time_two:
        h_bot = (Q*t)/(math.pi*r**2)
        h_top = h_inital
    else:
        h_bot = (Q*t)/(math.pi*r**2)
        h_top = h_inital-0.5*g*(t-time_two)**2-v_inital*(t-time_two)
    some_constant = (F**2)/(((math.pi*rho*orfice**2))**2)
    first_term = math.sqrt(some_constant + 2*g*(h_inital-h_bot))
    #print(first_term)
    second_term =  math.sqrt(some_constant + 2*g*(h_inital-h_top))
    y = (F/(2*g))*(first_term-second_term)
    mass2.append(y)

```

```python
plt.plot(time,mass2)
plt.ylabel('Mass (m)')
plt.xlabel('Time (s)')
plt.title('Mass (Region 2) vs Time')
```

```python
time = []
com2 = []
start_time = 0
end_time = (volume/Q)+(math.sqrt(v_inital**2+2*g*h_final)-v_inital)/g

time_one = (math.sqrt(v_inital**2+2*g*h_inital)-v_inital)/g
time_two = volume/Q

h_bot = 0
h_top = 0
time = np.arange(start_time, end_time, 0.001)
for t in time:
    if t <= time_one:
        h_bot = h_inital-0.5*g*t**2-v_inital*t
        h_top = h_inital
    elif t <= time_two:
        h_bot = (Q*t)/(math.pi*r**2)
        h_top = h_inital
    else:
        h_bot = (Q*t)/(math.pi*r**2)
        h_top = h_inital-0.5*g*(t-time_two)**2-v_inital*(t-time_two)
        
    some_constant = (F**2)/(((math.pi*rho*orfice**2))**2)
    first_term = math.sqrt(some_constant + 2*g*(h_inital-h_bot))
    second_term =  math.sqrt(some_constant + 2*g*(h_inital-h_top))
    mass_of_t = (F/(2*g))*(first_term-second_term)
    
    y= mass_of_t*-g
    
    com2.append(y)


```
